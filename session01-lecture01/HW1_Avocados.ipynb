```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
sns.set_style('whitegrid')
sns.set_palette('husl')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 11

print("Libraries loaded successfully! ✅")
```

## Load the Data

**Before you run code**: What questions would YOU ask before even seeing the data?

Think about:
- What time period does this cover?
- How granular is it? (daily? weekly? monthly?)
- What geographic coverage?
- Is it complete or are there gaps?

**This is critical**: Jumping straight into analysis without understanding your data leads to wrong conclusions.

```python
avocado = (
    pd.read_csv("../data/avocado.csv")
    .assign(
        type=lambda d: d["type"].astype("category"),
        region=lambda d: d["region"].astype("category"),
        date=lambda d: pd.to_datetime(d['date'], dayfirst=True)
    )
    .query('region != "TotalUS"')  # Remove aggregate to avoid double-counting
    .sort_values(['date', 'region', 'type'])
    .reset_index(drop=True)
)

# Add helpful derived columns
avocado['year'] = avocado['date'].dt.year
avocado['month'] = avocado['date'].dt.month
avocado['month_name'] = avocado['date'].dt.month_name()
avocado['quarter'] = avocado['date'].dt.quarter
avocado['week_of_year'] = avocado['date'].dt.isocalendar().week

print(f"✅ Data loaded: {avocado.shape[0]:,} rows × {avocado.shape[1]} columns")
print(f"📅 Date range: {avocado['date'].min().date()} to {avocado['date'].max().date()}")
print(f"🏪 Regions: {avocado['region'].nunique()}")
print(f"🥑 Types: {avocado['type'].unique().tolist()}")

avocado.head()
```
